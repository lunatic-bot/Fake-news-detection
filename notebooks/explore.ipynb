```python
from pathlib import Path

BASE_DIR = Path().resolve().parent
print(BASE_DIR)
```

```python
import pandas as pd
```

```python
fake_data_path = BASE_DIR / r'data\raw\Fake.csv'
fake_data = pd.read_csv(fake_data_path)
fake_data
```

```python
true_data_path = BASE_DIR / r'data\raw\true.csv'
true_data = pd.read_csv(true_data_path)
true_data
```

```python
fake_data.shape
```

```python
fake_data.dtypes
```

```python
any(fake_data['text'].isna())
```

```python
any(true_data['title'].isna())
```

```python
## add true lable as 1
true_data['label'] = 1
```

```python
## add fake lable as 0
fake_data['label'] = 0
```

```python
## concate both data
news_data = pd.concat([true_data, fake_data])
news_data
```

```python
## Suffle teh data 
df = news_data.sample(frac=1, random_state=42).reset_index(drop=True)
df
```

```python
df['full_text'] = df['title'].fillna('') + " " + df['text'].fillna('')
```

```python
new_df = df[['full_text', 'subject', 'label']]
```

```python
new_df
```

```python
csv_path = BASE_DIR / r'data\raw\combined_news_data.csv'
new_df.to_csv(csv_path, index=False)
```
